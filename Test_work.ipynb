На ферме содержатся шесть разных видов животных, и каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит всарай 6 раз.
Каково математическое ожидание количества разных видов животных, которые фермер увидит за день?


**Переведем исходные данные в математику:**
- 6 видов животных.
- фермер заходит в сарай 6 раз.
- каждый раз фермер видит случайное животное (вероятность увидеть любой вид одинакова и равна 1/6)

Нужно найти среднее количество разных видов, которые он увидит за день.

Обозначим:

X — случайная величина, равная количеству разных видов животных, увиденных за день;

E[X] — искомое математическое ожидание.

**Решение:**

1. Посчитаем вероятность не увидеть конкретный вид за 6 заходов:
- Вероятность не увидеть конкретный вид за один заход: 1−1/6=5/6 
- Вероятность не увидеть этот же вид за 6 заходов подряд: (5/6)^6 = 0.3349

2. Посчитаем вероятность увидеть конкретный вид хотя бы один раз
   - 1 - (5/6)^6 = 0,6651

3. Всего 6 видов, и каждый имеет вероятность 0,6651 быть увиденным. Чтобы найти среднее количество разных видов,  умножаем полученную вероятность на 6 * 0,6651= 3,9906.

Округляем: 3,99, то есть в среднем фермер увидит 3,99 разных видов животных за день.

**Задание 2: Кулинарное соревнование**

В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. Каково математическое ожидание числа победителей?


**Переведем исходные данные в математику:**
- Участвуют 80 шеф‑поваров с уникальными уровнями мастерства (от 1 — самый слабый до 80 — самый сильный).
- Этап 1: шефов случайным образом разбивают на 40 пар; в каждой паре побеждает более сильный шеф.
- Этап 2: снова случайное разбиение на 40 пар (пары могут повториться); снова побеждает более сильный.
- Победитель конкурса — тот, кто выиграл оба этапа.
- Предположим, что так как каждый мастер владеет уникальным уровнем навыка, то нет мастеров с одинаковым уровнем и все 80 мастеров имеют разный уровень и не будет 2 победителей в одном конкурсе.

Нужно найти среднее (математическое ожидание) число таких победителей.

**Решение:**

1. У каждого мастера может быть 79 соперников.
2. Конкретный шеф путь имеет уровень мастерства k (k принимает значение от 1...80).
3. Слабее него: k − 1 шефов.
4. Вероятность попасть на более слабого соперника и выиграть этап: (k-1) / 79
5. Всего 2 этапа отбобра и они независимы, то есть вероятностью победить в двух этапах: $$
P_{\text{обе\_победы}} = \left(\frac{k - 1}{79}\right)^2
$$
6. Чтобы найти среднее число победителей, нужно сложить вероятности победы для каждого шефа: $$
E = \sum_{k=1}^{79} \left[\left(\frac{k - 1}{79}\right)^2\right]
$$

7. $$
E = \frac{167\,320}{79^2} = \frac{167\,320}{6\,241} \approx 26{,}81
$$

Округляем до десятых: **26,8**.


**Задание 3: Одинокая дорога**

На пустынном шоссе вероятность появления автомобиля за 30-минутный период составляет 0.95.
Какова вероятность его появления за 10 минут? А за 27 минут?

**Переведем исходные данные в математику:**
- Вероятность увидеть машину за 30 минут: 0,95 (или 95 %).
- Нужно найти вероятность увидеть машину:
  - за 10 минут;
  - за 27 минут.

**Решение**
1. Найдём вероятность НЕ увидеть машину за 30 минут

Если вероятность увидеть машину — $0{,}95$, то вероятность **не** увидеть её: $$ 1 - 0{,}95 = 0{,}05 \quad (5\%).$$

2. Рассчитаем вероятность не увидеть машину за 10 минут

30 минут = три отрезка по 10 минут. Чтобы за 30 минут не было ни одной машины, она не должна появиться ни в первые 10 минут, ни во вторые, ни в третьи.

Обозначим:
* $X$ — вероятность увидеть машину за 10 минут;
* $1 - X$ — вероятность **не** увидеть машину за 10 минут.

Тогда:
$$
(1 - X)^3 = 0{,}05.
$$

Решаем уравнение:
1. Извлекаем кубический корень из обеих частей:
$$
1 - X = \sqrt[3]{0{,}05} \approx 0{,}3684.
$$
2. Находим $X$:
$$
X = 1 - 0{,}3684 = 0{,}6316.
$$

Округляем до сотых: **0,63** (или 63 %).

**Ответ для 10 минут:** вероятность увидеть машину — **63 %**.

3. Рассчитаем вероятность увидеть машину за 27 минут
Разбиваем 30 минут на 30 отрезков по 1 минуте.
Обозначим:
* $Q_1$ — вероятность не увидеть машину за 1 минуту.
Тогда:
$$
(Q_1)^{30} = Q_{30} = 0{,}05.
$$
Решаем уравнение:
$$
Q_1 = \sqrt[30]{0{,}05} = 0{,}05^{1/30}.
$$
Вычисляем:
$$
Q_1 \approx 0{,}9048.
$$
4. Найдём вероятность НЕ увидеть машину за 27 минут
27 минут = 27 отрезков по 1 минуте, поэтому:
$$
Q_{27} = (Q_1)^{27} \approx (0{,}9048)^{27}.
$$
Вычисляем:
$$
Q_{27} \approx 0{,}0789.
$$
5. Найдём вероятность увидеть машину за 27 минут
$$
P_{27} = 1 - Q_{27} \approx 1 - 0{,}0789 = 0{,}9211.
$$
Округляем до сотых: **0,92** (или 92 %).
**Ответ для 27 минут:** вероятность увидеть машину — **92 %**.

**Задание 1: Изоморфизмы**

Реализовать функцию (или тело функции), которая проверяет на изоморфность два слова. Пояснение: строки s и t называются изоморфными, если все вхождения каждого символа строки s можно последовательно заменить другим символом и получить строку t. Порядок символов при этом должен сохраняться, а замена — быть уникальной. Так, два разных символа строки s нельзя заменить одним и тем же символом из строки t, а вот одинаковые символы в строке s должны заменяться одним и тем же символом. Оценить оптимальность решения по времени и памяти и прикрепить текст кода.



```python
def edf(s: str, t: str):
    if len(s) != len(t):
        return False

    # Словари для хранения соответствий символов
    s_t = {} 
    t_s = {}  

    for i in range(len(s)):
        i_s, i_t = s[i], t[i]

        if i_s in s_t:
            if s_t[i_s] != i_t:
                return False
        else:
            s_t[i_s] = i_t

        # Проверяем соответствие
        if i_t in t_s:
            if t_s[i_t] != i_s:
                return False
        else:
            t_s[i_t] = i_s

    return True
```

- Алгоритм проходит по строкам один раз, где n — длина строк.
- Все операции со словарями (проверка наличия ключа, добавление элемента, получение значения) выполняются за O(1) в среднем случае.
- Таким образом, общее время выполнения пропорционально длине строк.

**Задание 2: Натуральная последовательность**

Реализовать функцию (или тело функции), которая находит единственное отсутствующее число из последовательности натуральных чисел 1,2,…,n.


```python

def missing_number(nums: list[int]) -> int:
    n = len(nums) + 1  # Общее количество чисел в полной последовательности
    expected_sum = n * (n + 1) // 2  # Сумма арифметической прогрессии
    actual_sum = sum(nums)  # Сумма имеющихся чисел
    return expected_sum - actual_sum
```

Линейная зависимость от размера входных данных.

**Задание 3: Факторизация**

Реализовать функцию (или тело функции), которая при введении натурального числа n разбивает его на простые множители (представить его в виде простых чисел).


```python
def factorize(n: int) -> list[int]:
    if n < 2:
        raise ValueError("Число должно быть натуральным и >= 2")
    factors = []
    divisor = 2  # Начинаем с наименьшего простого числа
    while divisor * divisor <= n:
        # Пока число делится на текущий делитель, добавляем его в результат
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    # Если после всех делений осталось число > 1, оно простое и является множителем
    if n > 1:
        factors.append(n)
    return factors
```

   * **Нельзя утверждать**, что выпитый кофе напрямую вызывает снижение баллов. Корреляция не подразумевает причинно‑следственную связь.
   * Возможные альтернативные объяснения:
     * **Стресс.** Студенты, испытывающие сильный стресс, могут пить больше кофе и при этом хуже справляться с экзаменом.
     * **Недостаток сна.** Те, кто мало спал перед экзаменом, пьют кофе для бодрости, но недосып влияет на результаты.
     * **Обратная связь.** Студенты с низкими ожиданиями от экзамена могут пить больше кофе «на всякий случай».
     * **Индивидуальные особенности.** Реакция на кофеин различается: у одних он улучшает концентрацию, у других — вызывает тревогу.
     * **Другие факторы.** На баллы влияют подготовка, сложность экзамена, здоровье студента и т. д.

3. **Рекомендации для дальнейшего анализа:**
   * Провести **контролируемый эксперимент** (например, рандомизированное распределение групп по потреблению кофе).
   * Собрать **дополнительные данные** (продолжительность сна, уровень стресса, время подготовки).
   * Проверить корреляцию на **большей выборке** для подтверждения устойчивости результата.

**Краткий итог:** между переменными есть сильная отрицательная корреляция ($r = -0{,}86$), но это не доказывает, что кофе снижает баллы. Для установления причинно‑следственной связи нужны дополнительные исследования.
